```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/drive/MyDrive/heart.csv')
df.head()
```

```python
corr =df.corr()
plt.figure(figsize=(18,10))
sns.heatmap(corr,annot=True)
plt.show()
```

```python
sns.countplot(x=df.target,hue='sex',data=df,palette=['#1CA53B','red'])
plt.xticks([0,1],labels=['Not Disease','Disease'])
plt.title("Target distriution")
```

```python
   plt.figure(figsize=(18,10))
   sns.countplot(x='age',hue='sex',data=df,palette=['#1CA53B','red'])
   plt.legend(["haven't Disease","have Disease"])
   plt.title('Heart disease frequency for ages')
   plt.xlabel('Age')
   plt.ylabel('Frequency')
   plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.distplot(df.age[df['target']==0],bins=30,color='red',label='have Disease')
sns.distplot(df.age[df['target']==1],bins=30,color='blue',label='havent Disease')
plt.legend()
plt.title('Heart Disease Distribution for Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
